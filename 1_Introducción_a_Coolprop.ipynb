# Introducción a Coolprop

Instalación de biblioteca CoolProp y cálculo de propiedades.

Primero, debemos instalar la biblioteca que vamos a utilizar para sacar las propiedades termodinámicas de los diferentes fluidos. Para esto, la librería que utilizaremos se llama Coolprop.

http://www.coolprop.org/

```python
pip install CoolProp
```

Para poder utilizar la librería, debemos primero importarla, Por ende:

```python
import CoolProp.CoolProp as cp
```

Coolprop admite una gran cantidad de fluidos puros con los cuales trabajar.

http://www.coolprop.org/v4/Fluids/pure_fluids.html

Para este ejemplo trabajaremos con Agua líquida saturada como el fluido de trabajo. Definiremos como la temperatura de referencia 30 °C y una calidad de referencia de 0. Según las tablas de propiedades, obtendríamos los siguientes valores: 
image.png

  Para utilizar Coolprop, primero debemos definir nuestras dos variables de entrada:

```python
fluid = 'water'
T_ref = 30 + 273.15 # K
Q_ref = 0
```

Calculemos la densidad del agua a la temperatura y presión de referencia. Las variables a utilizar por Coolprop deben ingresarse con las unidades adecuadas, tal como se muestra en la tabla a continuación:

image.png

La forma en como llamaremos a las propiedades del agua a las condiciones de referencia previamente establecidas, será de la siguiente forma:

cp.PropsSI(in1, in2, in3=None, in4=None, in5=None, in6=None, in7=None)

De esta forma, primera obtendremos la densidad del agua

```python
density = cp.PropsSI("D", "T", T_ref, "Q", Q_ref, fluid) # (kg/m3)
density
```

La entalpía sería:

```python
enthalpy = cp.PropsSI("H", "T", T_ref, "Q", Q_ref, fluid) # (J/kg)
enthalpy
```

La entropía sería:

```python
entropy = cp.PropsSI("S", "T", T_ref, "Q", Q_ref, fluid) # (J/kg)
entropy
```

La energía interna sería:

```python
int_energy = cp.PropsSI("U", "T", T_ref, "Q", Q_ref, fluid) # (J/kg)
int_energy
```

El volumen específico es:

```python
v_esp = 1 / cp.PropsSI("D", "T", T_ref, "Q", Q_ref, fluid)  # m3/kg
v_esp
```

Si quisiermos ahora evaluar las propiedades de algún otro fluido, Coolprop admite una gran cantidad de fluidos puros, los cuales pueden verse en el siguiente link:

https://coolprop.sourceforge.net/fluid_properties/PurePseudoPure.html#list-of-fluids

Ahora, que tal si no conocemos la fase en la cual se encuentra el agua, pero sabemos que se encuentra a una temperatura de 300 K y una presión de 120 kPa. Entonces, para conocer en que fase se encuentra podemos utilizar el siguiente código:

```python
fluid = "water"
T_ref = 300 # K
P_ref = 120000 # Pa

phase = cp.PhaseSI("T", T_ref, "P", P_ref, fluid)
phase
```
